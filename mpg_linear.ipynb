# Dataset : mpg (miles per gallon)

<span>The csv file contains data regrading car and target variable is <b>mpg(miles per gallon).</b> The features are cylinders,displacement,horsepower,weight,acceleration,model year,origin and car name </span>

```python
import pandas as pd

df = pd.read_csv('auto-mpg[1].csv')
df.head()
```

### <span style="color:red">Ecoding the string values to numeric</span>

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['car name'] = label_encoder.fit_transform(df['car name'])
df['horsepower'] = label_encoder.fit_transform(df['horsepower'])
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
X= df.drop('mpg',axis=1)
X.head()
```

```python
y=df['mpg']
y.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(X.corr(),annot=True)
plt.title('Feature Correlation Heatmap')
plt.show()

```

### <span style="color:red">Linear Regression</span>

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=42)
X_train.shape
```

```python
X_test.shape
```

```python
reg_lin=LinearRegression()
reg_lin.fit(X_train,y_train)
```

```python
y_pred= reg_lin.predict(X_test)
```

### <span style="color:red">Cost and Accuracy of Linear Regression</span>

```python
mse = mean_squared_error(y_test,y_pred)
print(mse)
```

```python
r2 = r2_score(y_test,y_pred)
print(r2)
```

### <span style="color:red">Polynomial Regression</span>

```python
from sklearn.preprocessing import PolynomialFeatures
reg_poly=PolynomialFeatures(degree=2)
X_train_poly = reg_poly.fit_transform(X_train)
X_test_poly = reg_poly.fit_transform(X_test)
```

```python
lr_poly = LinearRegression()
lr_poly.fit(X_train_poly,y_train)
```

```python
y_poly_pred = lr_poly.predict(X_test_poly)
```

### <span style="color:red">Cost and Accuracy of Polynomial Regression</span>

```python
mse_poly= mean_squared_error(y_test,y_poly_pred)
print(mse_poly)
```

```python
r2_poly=r2_score(y_test,y_poly_pred)
print(r2_poly)
```
